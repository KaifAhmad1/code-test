```python
import torch
import time
import psutil
from diffusers import CogVideoXPipeline
from diffusers.utils import export_to_video
from diffusers import BitsAndBytesConfig
```

```python
# Prompt and negative prompt
PROMPT = ("A sleek, modern smartphone with a glossy black finish rests on a polished wooden table in a brightly lit showroom. "
          "The camera slowly pans around the phone, highlighting its slim design and reflective surface. "
          "A soft, natural light from a nearby window casts gentle shadows, enhancing the phone’s premium look. "
          "In the background, a blurred display screen shows vibrant app icons, subtly pulsing with color. "
          "A person’s hand, with neatly manicured nails, gently picks up the phone, tilts it to showcase its edge-to-edge screen, "
          "and places it back down. The scene feels lively yet professional, with smooth motion and realistic textures, "
          "resembling high-quality commercial footage.")
NEGATIVE_PROMPT = ("Blurry, low resolution, distorted, jittery motion, inconsistent lighting, unrealistic textures, "
                   "static scenes, disconnected objects, low-quality rendering, cartoonish, overexposed, underexposed, "
                   "artifacts, unnatural colors, poor frame coherence.")
```

```python
def get_memory_usage():
    """Return GPU and system memory usage in GB."""
    if torch.cuda.is_available():
        gpu_memory = torch.cuda.memory_allocated() / 1024**3
        gpu_memory_reserved = torch.cuda.memory_reserved() / 1024**3
    else:
        gpu_memory = gpu_memory_reserved = 0
    system_memory = psutil.virtual_memory().used / 1024**3
    return gpu_memory, gpu_memory_reserved, system_memory
```

```python
def main():
    try:
        print("Initializing CogVideoX pipeline...")
        start_time = time.time()

        # Quantization config
        quant_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_compute_dtype=torch.bfloat16)

        # Load pipeline
        pipe = CogVideoXPipeline.from_pretrained(
            "THUDM/CogVideoX-5b",
            torch_dtype=torch.bfloat16,
            quantization_config=quant_config
        )

        # Apply memory optimizations
        print("Applying memory optimizations: VAE tiling, CPU offloading...")
        pipe.vae.enable_tiling()
        pipe.enable_model_cpu_offload()
        pipe.enable_sequential_cpu_offload()

        gpu_mem, gpu_res, sys_mem = get_memory_usage()
        print(f"Pipeline loaded. Memory usage: GPU {gpu_mem:.2f} GB (Reserved {gpu_res:.2f} GB), System {sys_mem:.2f} GB")

        # Generate video
        print("Generating video with CogVideoX...")
        video = pipe(
            prompt=PROMPT,
            negative_prompt=NEGATIVE_PROMPT,
            height=480,
            width=720,
            num_frames=49,
            num_inference_steps=50,
            guidance_scale=6.0,
            generator=torch.manual_seed(42)
        ).frames[0]

        # Export video
        print("Exporting video to cogvideox_output.mp4...")
        export_to_video(video, "cogvideox_output.mp4", fps=8)
        end_time = time.time()

        gpu_mem, gpu_res, sys_mem = get_memory_usage()
        print(f"Video generated: cogvideox_output.mp4")
        print(f"Inference time: {end_time - start_time:.2f} seconds")
        print(f"Final memory usage: GPU {gpu_mem:.2f} GB (Reserved {gpu_res:.2f} GB), System {sys_mem:.2f} GB")

    except Exception as e:
        print(f"Error in CogVideoX pipeline: {str(e)}")
        raise

if __name__ == "__main__":
    main()
```
